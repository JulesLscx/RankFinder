```python
from sentence_transformers import SentenceTransformer
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow import keras
from tensorflow.keras import layers
import tensorflow as tf
```

```python
PATH = '../data/Sample_Books_rating.csv'

def load_data():
    df = pd.read_csv(PATH)
    return df
```

```python
data = load_data()
```

```python
X = data["revue/texte"]
```

```python
y = data["revue/score"]
```

```python
y_polarite = y.apply(lambda x: 2 if x > 3 else 1 if x == 3 else 0)
```

```python
y_polarite.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
model2 = SentenceTransformer('paraphrase-mpnet-base-v2')

```

```python
import tqdm
```

```python
X_train = X_train.apply(lambda x: model2.encode(x))
```

```python
X_test = X_test.apply(lambda x: model2.encode(x))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=27)
forest = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
knn.fit(X_train.tolist(), y_train)
```

```python
forest.fit(X_train.tolist(), y_train)
```

```python
knn.score(X_test.tolist(), y_test)
```

```python
forest.score(X_test.tolist(), y_test)
```
